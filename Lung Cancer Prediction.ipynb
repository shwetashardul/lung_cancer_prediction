# Preprocessing

## Encoding

```python
# Define the order of the categories
levels = ['Low', 'Medium', 'High']

# Create a dictionary to map categories to numerical values
level_mapping = {size: index for index, size in enumerate(levels)}

# Apply the ordinal encoding to the 'Size' column
df['Level_encoded'] = df['Level'].map(level_mapping)
```

```python
level_mapping
```

## Feature Selection

```python
cols=np.array(['Age', 'Gender', 'Air Pollution', 'Alcohol use', 'Dust Allergy', 'OccuPational Hazards', 'Genetic Risk', 'chronic Lung Disease', 'Balanced Diet', 'Obesity', 'Smoking', 'Passive Smoker', 'Chest Pain', 'Coughing of Blood', 'Fatigue', 'Weight Loss', 'Shortness of Breath', 'Wheezing', 'Swallowing Difficulty', 'Clubbing of Finger Nails', 'Frequent Cold', 'Dry Cough', 'Snoring'])
temp_X=df[cols]
temp_y=df['Level']
```

```python
import numpy as np
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2


# Create SelectKBest instance with chi-square as the score function
k_best = SelectKBest(score_func=chi2, k=4)

# Fit and transform the features using SelectKBest
X_k_best = k_best.fit_transform(temp_X, temp_y)

# Get the indices of the selected features
selected_features_indices = k_best.get_support(indices=True)

# Print the selected features
tr_cols=cols[selected_features_indices]
tr_cols
```

```python
import numpy as np
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import mutual_info_classif


# Create SelectKBest instance with chi-square as the score function
k_best = SelectKBest(score_func=mutual_info_classif, k=4)

# Fit and transform the features using SelectKBest
X_k_best = k_best.fit_transform(temp_X, temp_y)

# Get the indices of the selected features
selected_features_indices = k_best.get_support(indices=True)

# Print the selected features
tr_cols=cols[selected_features_indices]
tr_cols
```

# Model Training

## Split the Data

```python
from sklearn.model_selection import train_test_split
# Define the percentage of data to be used for testing (e.g., 20% for testing)
test_size = 0.2
X=df[tr_cols]
y=df['Level_encoded']
# Set a random seed for reproducibility (optional but recommended)
random_seed = 42

# Perform the train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=random_seed)
```

## Building the models

```python
from sklearn.svm import SVC

svm_classifier = SVC(kernel='linear', C=1.0)

# Train the SVM classifier on the training data
svm_classifier.fit(X_train, y_train)
# Get the hyperplane parameters
w = np.round(svm_classifier.coef_[0],2)
b = np.round(svm_classifier.intercept_[0],2)
print(w,b)
```

```python
from sklearn.tree import DecisionTreeClassifier
# Create a Decision Tree classifier
dt_clf = DecisionTreeClassifier(max_depth=5)

# Train the Decision Tree classifier on the training data
dt_clf.fit(X_train, y_train)
```

```python
from sklearn.tree import export_graphviz
from six import StringIO
import pydot
import matplotlib.pyplot as plt
from IPython.display import Image

# This function creates images of tree models using pydot
def print_tree(estimator, features, class_names=None, filled=True):
    tree = estimator
    names = features
    color = filled
    classn = class_names

    dot_data = StringIO()
    export_graphviz(estimator, out_file=dot_data, feature_names=features, class_names=classn, filled=filled)
    graph = pydot.graph_from_dot_data(dot_data.getvalue())
    return(graph)

graph, = print_tree(dt_clf,X_train.columns)
Image(graph.create_png())
```

```python
from sklearn.linear_model import LogisticRegression
# Create a logistic regression model
lr_clf = LogisticRegression()

# Train the model on the training data
lr_clf.fit(X_train, y_train)
```

# Evaluation

## SVM

```python
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# Make predictions on the test data
y_pred = svm_classifier.predict(X_test)

# Assuming you have true labels and predicted labels
true_labels = y_test
predicted_labels = y_pred

# Generate confusion matrix
cm = confusion_matrix(true_labels, predicted_labels)

# Plot the confusion matrix using seaborn and matplotlib
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=["Class 0", "Class 1", "Class 2"], yticklabels=["Class 0", "Class 1", "Class 2"])
plt.title("Confusion Matrix: SVM")
plt.xlabel("Predicted Labels")
plt.ylabel("True Labels")
plt.show()
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
# Evaluate the classifier's performance (e.g., using accuracy)
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

# Calculate the F1 score
f1 = f1_score(y_test, y_pred, average='weighted')  # 'weighted' accounts for class imbalance
print("F1 Score:", f1)
```
